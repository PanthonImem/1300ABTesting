```python
df = pd.read_csv('myfilteredlog_raw.csv')
df.columns = ['key','drop1','drop2','page','ts_pageload','ts_action','action','user']
df = df.drop(['drop1','drop2'], axis = 1)

print('# Unique Users', df['user'].nunique())
pd.set_option('display.max_rows', None)
df.sort_values(['user','ts_pageload'])
```

```python
df['return'] = 0
df = df.sort_values(['user','ts_pageload']).reset_index(drop = True)
for i in range(1,df.shape[0]):
    if((df.iloc[i]['user'] == df.iloc[i-1]['user'])
      and ((df.iloc[i-1]['page']=='C')|
           (df.iloc[i-1]['action']!=0))
       and df.iloc[i]['ts_pageload']!=df.iloc[i-1]['ts_pageload']
       and df.iloc[i]['page']!='C'
      ):
        df.at[i,'return'] = 1
df
gdf = df.groupby(['user','page'])['return'].sum().to_frame().reset_index()
gdf = gdf[gdf['page']!='C'].reset_index()
#drop my own repeated visit from page A result
gdf = gdf.drop(35, axis = 0)
gdf.loc[gdf['return']>1,'return'] = 1
ggdf = gdf.groupby('page').agg({'return':'sum','user':'nunique'}).reset_index()
ggdf.columns = ['page','return','not_return']
ggdf['not_return'] = ggdf['not_return'] - ggdf['return']
ggdf
```

```python
"""
Data Processing steps
1) We want to first get return tag before we clean non-session pageid
1) clean pageload with no action
2) clean users who visited and do task multiple times, keep only the first

"""
```

```python
#Now that we are finished with return rate, we want to only keep pageload with timestamp for calculating task time
```

```python
df['keep'] = 0
pageload_keep = df[(df['page']=='C')|(df['action']!='0')]['ts_pageload'].unique()

df.loc[df['ts_pageload'].isin(pageload_keep),'keep'] = 1
cdf = df[df['keep']==1]

keepts = cdf.drop_duplicates(subset = ['user','ts_pageload'], keep = 'first')['ts_pageload'].unique()
cdf = cdf[cdf['ts_pageload'].isin(keepts)].reset_index(drop = True)
cdf
```

```python
cdf['task_time'] = cdf['ts_action'] - cdf['ts_pageload']
for i in range(cdf.shape[0]): 
    if(cdf.iloc[i]['page']=='C'):
        cdf.at[i,'task_time'] = cdf.iloc[i]['ts_pageload'] - cdf.iloc[i-1]['ts_pageload'] 
gggdf = cdf.groupby(['user','page'])['task_time'].max().to_frame().reset_index()


ggggdf = gggdf.groupby(['user'])['task_time'].max().to_frame().reset_index().merge(gggdf[gggdf['page']!='C'][['user','page']].drop_duplicates(), how = 'left', left_on = 'user',right_on = 'user')
ggggdf
print('A:',list(ggggdf[ggggdf['page']=='A']['task_time'].unique()))
print('B:',list(ggggdf[ggggdf['page']=='B']['task_time'].unique()))
```
